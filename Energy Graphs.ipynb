### CREATE A CLASS FOR ENERGY DATA VISUALIZATION

```python
import pandas as pd 
import math
import matplotlib.pyplot as plt
#create class energy
class Energy():
    def __init__(self):
        print('Open your energy grah')
      
    
    
    
    def file_select(self, filename):
        self.filename = filename
        self.df = pd.read_csv(filename)
        print(self.df)
  
        


    #Create attribute day 
    def average_day_energy(self):
        ax= self.df.plot(kind='bar', x='START TIME', y='USAGE')
        plt.title('Whole year average energy usage for one day')
        ax.set_ylabel('KWH')
        plt.show()
    
   


    #create a month attribute
    def one_months_data(self):        
        df2=  pd.DataFrame({
            'usage':[],
            'date':[]  
        })
        userinput =input('Input index \n')
        i = int(userinput)
        if i > len(self.df)-720:
            print('That value is to high')
            userinput =input('Input index \n')
            i = int(userinput)
        index = self.df.loc[int(userinput)]
        use = index["USAGE"]
        date = index["DATE"]      
        for i in range (i-2, i+720):  
            index = self.df.loc[i]
            use = index["USAGE"]
            date = index["DATE"]  
            df2 = df2.append({'usage' :use, 'date' : date} , ignore_index=True)
        #graph months worth
        ax = df2.plot(kind='line',x='date',y='usage')
        ax.set_xlabel("30 day time period")
        ax.set_ylabel('KWH')
        plt.rcParams['figure.figsize'] =(20,10)
        plt.show()     
   







    #create attribute compare
    def compare_month(self):
        ax= plt.gca()
        df2=  pd.DataFrame({
            'usage':[],
            'date':[]  
        })

        userinput =input('Input index \n')

        i = int(userinput)
        if i > len(self.df)-720:
            print('That value is to high')
            userinput =input('Input index \n')
            i = int(userinput)

        index = self.df.loc[int(userinput)]
        use = index["USAGE"]
        date = index["DATE"]      


        for i in range (i-2, i+720):
    
            index = self.df.loc[i]
            use = index["USAGE"]
            date = index["DATE"]  
            df2 = df2.append({'usage' :use, 'date' : date} , ignore_index=True)



        #graph months worth
        ax = df2.plot(kind='line',x='date',y='usage',ax=ax)
        ax.set_xlabel("30 day time period")
        ax.set_ylabel('KWH')
        
        
        df2=  pd.DataFrame({
            'usage':[],
            'date':[]  
        })

        userinput =input('Input index \n')

        i = int(userinput)
        if i > len(self.df)-720:
            print('That value is to high')
            userinput =input('Input index \n')
            i = int(userinput)

        index = self.df.loc[int(userinput)]
        use = index["USAGE"]
        date = index["DATE"]      


        for i in range (i-2, i+720):
    
            index = self.df.loc[i]
            use = index["USAGE"]
            date = index["DATE"]  
            df2 = df2.append({'usage' :use, 'date' : date} , ignore_index=True)
        

        #graph months worth
        ax = df2.plot(kind='line',x='date',y='usage', ax=ax)
        ax.set_xlabel("30 day time period")
        ax.set_ylabel('KWH')
        ax
        plt.rcParams['figure.figsize'] =(20,10)
        plt.show()   
        
        
        
        
        #create day attribute
    def one_day(self):        
        df2=  pd.DataFrame({
            'usage':[],
            'date':[]  
        })
        userinput =input('Input index \n')
        i = int(userinput)
        if i > len(self.df)-24:
            print('That value is to high')
            userinput =input('Input index \n')
            i = int(userinput)
        index = self.df.loc[int(userinput)]
        use = index["USAGE"]
        date = index["DATE"]      
        for i in range (i, i+24):  
            index = self.df.loc[i]
            use = index["USAGE"]
            date = index["DATE"]  
            df2 = df2.append({'usage' :use, 'date' : date} , ignore_index=True)
        #graph months worth
        ax = df2.plot(kind='line',x='date',y='usage')
        ax.set_xlabel("24hr time period")
        ax.set_ylabel('KWH')
        plt.rcParams['figure.figsize'] =(24,10)
        plt.show()        
    
        

    def compare_day(self):

        ax= plt.gca()
        df2=  pd.DataFrame({
            'usage':[],
            'date':[]  
        })

        userinput =input('Input index \n')

        i = int(userinput)
        if i > len(self.df)-720:
            print('That value is to high')
            userinput =input('Input index \n')
            i = int(userinput)

        index = self.df.loc[int(userinput)]
        use = index["USAGE"]
        date = index["DATE"]      


        for i in range (i, i+24):
    
            index = self.df.loc[i]
            use = index["USAGE"]
            date = index["DATE"]  
            df2 = df2.append({'usage' :use, 'date' : date} , ignore_index=True)



        #graph months worth
        ax = df2.plot(kind='line',x='date',y='usage',ax=ax)
        ax.set_xlabel("30 day time period")
        ax.set_ylabel('KWH')
        
        
        df2=  pd.DataFrame({
            'usage':[],
            'date':[]  
        })

        userinput =input('Input index \n')

        i = int(userinput)
        if i > len(self.df)-720:
            print('That value is to high')
            userinput =input('Input index \n')
            i = int(userinput)

        index = self.df.loc[int(userinput)]
        use = index["USAGE"]
        date = index["DATE"]      


        for i in range (i, i+24):
    
            index = self.df.loc[i]
            use = index["USAGE"]
            date = index["DATE"]  
            df2 = df2.append({'usage' :use, 'date' : date} , ignore_index=True)
        

        #graph months worth
        ax = df2.plot(kind='line',x='date',y='usage', ax=ax)
        ax.set_xlabel("24hr time period")
        ax.set_ylabel('KWH')
        ax
        plt.rcParams['figure.figsize'] =(20,10)
        plt.show()   




e= Energy()
e.file_select("/Users/kintienmalik/Documents/Stanford/energydata.csv")
```
